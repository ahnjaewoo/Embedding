## Speed Up

```python
# train_time dbpedia
num_workers = np.array([4, 8, 12, 16])

dbpedia_20_50 = np.array([9621.885369, 6483.7345, 5616.832994, 5895.937505])
dbpedia_10_100 = np.array([8638.436743, 5072.383917, 4278.821042, 4271.674929])
dbpedia_10_50 = np.array([4202.04374, 3344.924949, 2922.301889, 3003.355501])
dbpedia_baseline = np.array([5206.37, 3002, 1940.23, 1837.61])
dbpedia_single = 12490.4

fig, ax = plt.subplots(1, 1, figsize = (12, 9))

ax.plot(num_workers, dbpedia_single / dbpedia_20_50, marker = 'o', linestyle = '-', label = '20 / 50',
        lw = 4, markersize = 12, color = '#D4722F')
ax.plot(num_workers, dbpedia_single / dbpedia_10_100, marker = 'v', linestyle = '-', label = '10 / 100',
        lw = 4, markersize = 12, color = '#70AD47')
ax.plot(num_workers, dbpedia_single / dbpedia_10_50, marker = 'd', linestyle = '-', label = '10 / 50',
        lw = 4, markersize = 12, color = '#548BBD')
ax.plot(num_workers, dbpedia_single / dbpedia_baseline, marker = 's', linestyle = '-', label = 'paragraphE',
        lw = 4, markersize = 12, color = '#A5A5A5')
ax.plot([1, 16], [1, 16], c = 'k', linestyle = '--', label = 'Ideal', lw = 4, zorder = -1)
ax.set_xlim(0, 17)
ax.set_ylim(0, 17)
ax.set_xticks(num_workers)
ax.set_yticks(num_workers)
ax.set_xlabel('Number of workers')
ax.set_ylabel('Speed up')
legend = ax.legend(loc = 'upper center', bbox_to_anchor = (0.5, -0.18), fancybox = True, ncol = 3)

#plt.tight_layout()
plt.savefig('./line_speedup_transe_dbpedia.pdf', dpi = 300, quality = 100, bbox_inches = 'tight')
plt.show()
```
