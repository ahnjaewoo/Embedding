## transE - fb15k / wn18

### network 시각화

```python
fb15k_entities = list()
wn18_entities = list()
fb15k_node_4 = list()
wn18_node_4 = list()
fb15k_node_8 = list()
wn18_node_8 = list()
fb15k_edge = list()
wn18_edge = list()

colors = ['black', 'gray', 'red', 'sienna', 'gold', 'olivedrab', 'seagreen', 'mediumpurple', 'palevioletred']

def entity2worker(entity, maxmin_result):
    
    for idx, val in enumerate(maxmin_result):
        
        if entity in val:
            
            worker = idx
            break
    
    else:
        
        worker = len(maxmin_result)
        
    return worker

with open('./fb15k/maxmin_result_fb15k_4.txt', 'r') as fs:
    
    maxmin_fb15k_4 = list(map(eval, fs.readlines()))
    
with open('./fb15k/maxmin_result_fb15k_8.txt', 'r') as fs:
    
    maxmin_fb15k_8 = list(map(eval, fs.readlines()))
    
with open('./fb15k/train.txt', 'r') as fs:
    
    train_fb15k = fs.readlines()
    
with open('./wn18/maxmin_result_wn18_4.txt', 'r') as fs:
    
    maxmin_wn18_4 = list(map(eval, fs.readlines()))
    
with open('./wn18/maxmin_result_wn18_8.txt', 'r') as fs:
    
    maxmin_wn18_8 = list(map(eval, fs.readlines()))
    
with open('./wn18/train.txt', 'r') as fs:
    
    train_wn18 = fs.readlines()
    
for l in train_fb15k:
    
    triple = l.replace('\n', '').split('\t')
    fb15k_entities.append(triple[0])
    fb15k_entities.append(triple[2])
    
for l in train_wn18:
    
    triple = l.replace('\n', '').split('\t')
    wn18_entities.append(triple[0])
    wn18_entities.append(triple[2])

fb15k_entities = list(set(fb15k_entities))
wn18_entities = list(set(wn18_entities))
    
fb15k_entity2id = {fb15k_entities[idx]:idx for idx in range(len(fb15k_entities))}
wn18_entity2id = {wn18_entities[idx]:idx for idx in range(len(wn18_entities))}

for e in fb15k_entities:
    
    fb15k_node_4.append((fb15k_entity2id[e], {'color' : colors[entity2worker(e, maxmin_fb15k_4)]}))
    fb15k_node_8.append((fb15k_entity2id[e], {'color' : colors[entity2worker(e, maxmin_fb15k_8)]}))

for e in wn18_entities:
    
    wn18_node_4.append((wn18_entity2id[e], {'color' : colors[entity2worker(e, maxmin_wn18_4)]}))
    wn18_node_8.append((wn18_entity2id[e], {'color' : colors[entity2worker(e, maxmin_wn18_8)]}))

for l in train_fb15k:
    
    triple = l.replace('\n', '').split('\t')
    fb15k_edge.append((fb15k_entity2id[triple[0]], fb15k_entity2id[triple[2]]))

for l in train_wn18:
    
    triple = l.replace('\n', '').split('\t')
    wn18_edge.append((wn18_entity2id[triple[0]], wn18_entity2id[triple[2]]))
```
